```python
from sklearn import datasets 
import pandas as pd
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

diabetesdata = datasets.load_diabetes()         # Loading the diabetes dataset from sklearn sample dataset

diabetesdata
```

```python
x2 = diabetesdata.data                  # Putting the input data into x2 
y2 = diabetesdata.target                # Putting the output data into y2
```

```python
X2 = pd.DataFrame(x2)      # Converting the input data into dataframe for better understanding 

X2
```

```python
feature_names = diabetesdata.feature_names          # Getting the actual column names of dataset

feature_names
```

```python
X2.columns = feature_names                          # Assigning the actual column names to the dataset

X2
```

```python
diabetesdata.DESCR                         # Getting to know what each column stands for by knowing about the dataset in depth
```

```python
# Splitting the input and output data into training and testing data

x_train, x_test, y_train, y_test = model_selection.train_test_split(x2, y2)   
```

```python
# Creating a model by using linear regression algorithm and passing training data to train it

newalgo = LinearRegression()

newalgo.fit(x_train, y_train)
```

```python
# Passing testing data's input to the algorithm to get the predicted output

y_predicted = newalgo.predict(x_test)
```

```python
# Comparing predicted and actual output 

plt.scatter(y_predicted, y_test)
plt.xlabel('Predicted Output')
plt.ylabel('Actual Output')
plt.grid()
plt.show()
```
